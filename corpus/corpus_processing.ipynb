## Create randomly sampled files
for experimental mock-up

```python
df.WhPhaseType.values
```

```python
critical = df[(df['QuestionType'] == 'root')
              & 
              (df['DegreeQ'] == 'no' )
              &
              (df['Wh'].isin(['how','How','where','Where','who','Who']))
              &
              (df['WhPhaseType'] == 'monomorphemic')
             ]
```

```python
len(critical)
```

```python
df_sample = critical[["Item_ID","Sentence","Prev-Context"]].rename(columns={"Item_ID":"TGrepID","Sentence":"EntireSentence","Prev-Context":"PreceedingContext"})
```

```python
d = df_sample.sample(30, random_state=321)
```

```python
def paramap(df):
    if df["EntireSentence"].str.contains(["who|Who"]):
        return "people"
    if df["EntireSentence"].str.contains(["where|Where"]):
        return "places"
    if df["EntireSentence"].str.contains(["how|How"]):
        return "ways"
```

```python
d["AResponse"] = "What is a way?"
d["SomeResponse"] = "What are some of the ways?"
d["AllResponse"] = "What are all of the ways?"
d["TheResponse"] = "What is the way?"
```

```python
d.to_csv("testfile.txt",header=True,sep="\t")
```
